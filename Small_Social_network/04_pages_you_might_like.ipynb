```python
'''
On social media platforms, users interact with pages by liking, following, 
or engaging with posts. The goal is to analyze these interactions and suggest 
relevant pages based on user behavior.
'''
```

```python
# loading data
import json
data = json.load(open("03_data.json"))

# let's suppose we need to find out suggested pages for id no
user_id = 5

# pages liked by user
liked_pages = set(data['users'][user_id - 1]['liked_pages'])

# making a list of other users liked pages that have some mutual liking
suggested_pages = []
for user in data['users']:
    if user['id'] != user_id:
        a = set(user['liked_pages'])
        count = len(liked_pages.intersection(a))
        if count > 0:
            suggestion = list(a - liked_pages.intersection(a))
            suggested_pages.append(suggestion)

# making a dictionary of suggested page as key and its preference as value
l = []
unique = {}

for i in suggested_pages:
    l.append(i[0])
for i in l:
    count = l.count(i)
    unique[i] = count

# sorting the suggested pages with most suggested first and then converting back to list
sorted_unique = dict(sorted(unique.items(), key=lambda item:item[1], reverse = True))
sorted_unique_list = list(sorted_unique.keys())

# Printing out the recommedation
print("The suggested pages that you might like in the most probable order are:")
for i in sorted_unique_list:
    a = data["pages"][i-101]
    print(f"ID {a["id"]} : {a["name"]}")
```
