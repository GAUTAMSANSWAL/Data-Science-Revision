```python
'''
Build a 'People You May Know' feature!
In social networks, this feature helps users connect with others by suggesting friends 
based on mutual connections. Your job is to analyze mutual friends and recommend potential 
connections.
'''
```

```python
# loading data
import json
data = json.load(open("03_data.json"))

# let's say we need to find out people we know for id no 1
username_id = 1

# lets find out direct friends first
direct_friends = data["users"][username_id - 1]['friends']

# now we need to find the friends of friends
fof = []
for uid in direct_friends:
    mutuals = [friend for friend in data["users"][uid - 1]['friends']]
    for i in mutuals:
        if i!=username_id and i not in direct_friends:
            fof.append(i)
fof.sort()

# storing the suggested people with their degree of knowing in a dict
unique_list = []
mutual_dict = {}
for i in fof:
    if i not in unique_list:
        unique_list.append(i)
        a = fof.count(i)
        mutual_dict[i] = a
    else:
        continue

# sorting the dictionary values in descending orders to make a list of suggested people 
sorted_dict = dict(sorted(mutual_dict.items(), key=lambda item:item[1], reverse=True))

# Printing the ID and names of the people you may know
pymk_id = list(sorted_dict.keys())
print("This is the list of the people you may know")
for i in pymk_id:
    name = data["users"][i - 1]['name']
    print(f"{name} (ID: {i}) ")
```
