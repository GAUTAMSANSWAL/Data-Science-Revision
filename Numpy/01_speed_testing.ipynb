```python
import numpy as np
import time
```

```python
size = 1_000_000
l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))
```

```python
start = time.time()
sum = l1 + l2
end = time.time()
print(end - start)
```

```python
size = 1_000_000
l01 = list(range(size))
l02 = list(range(size))
```

```python
start = time.time()
sum = [a+b for a,b in zip(l01, l02)]
end = time.time()
print(end - start)
```
