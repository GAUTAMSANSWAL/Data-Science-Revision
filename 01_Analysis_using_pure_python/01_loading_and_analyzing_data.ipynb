```python
import json
```

```python
# writing a function to access json file

def load_data(file):
    with open(file, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("Project1/data.json")
```

```python
def display_data(data):
    for user in data['users']:
        print(f"{user['name']} is friends with ID no {user['friends']}")
    print("\n")
    for user in data['users']:
        print(f"{user['name']} likes page ID no {user['liked_pages']}")
    print("\n")
    for page in data['pages']:
        print(f"{page['id']} : {page['name']}")
    print("\n")
```

```python
display_data(data)
```
