```python
import matplotlib.pyplot as plt
```

```python
# Data
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
explode = [0 , 0.3 , 0 , 0]
plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels, explode=explode, shadow=True, autopct = '%1.2f%%')
```

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    shadow = True,
    startangle = 250,
    wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': ':'}
)
plt.show()
```
