```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3], [4, 5, 6])
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
plt.xlabel('Years')
plt.ylabel('Runs')
plt.title('Sachin tendulkar yearly runs')
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label='Kohli score')
plt.plot(years, sehwag, label='Sehwag score')
plt.xlabel('Years')
plt.ylabel('Runs')
plt.title('Kohli vs Sehwag Performance')
plt.legend()
```

```python
plt.plot(years, kohli,'go--', label='Kohli score')
plt.plot(years, sehwag, 'r^:' ,label='Sehwag score')
```

```python
plt.plot(years, kohli,color='red',linestyle='--' ,label='Kohli score')
plt.plot(years, sehwag, color='purple', linestyle='-.' ,label='Sehwag score')
plt.plot(years, runs, color='green', linestyle=':', linewidth = '2',label='Tendulkar score')
plt.legend()
plt.xlabel('Years')
plt.ylabel('Runs')
plt.title('Tendulkar vs Sehwag vs Kohli')
plt.grid(True)
plt.style.use('seaborn-v0_8-pastel')
```
