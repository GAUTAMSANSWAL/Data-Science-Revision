```python
# Opening raw text file data
with open("pre_data.txt", encoding = "utf-8") as f:
    data = f.read()
```

```python
# Converting grouped data in individual chunks
chunks = data.split("\n\n")
```

```python
chunks[0]
```

```python
# Seprating artibutes from individual chunks, making a dictionary of it and containing all in a list
organized_data = []
temp = {}
i = 1
for chunk in chunks:
    lines = chunk.strip().split("\n")
    username = lines[0]
    followers = float(lines[1].split(' follower')[0].replace('M', '').replace('K', ''))
    if ('K' in lines[1]):
        followers = int(followers*1000)
    elif ('M' in lines[1]):
        followers = int(followers*1000000)
    else:
        followers = int(followers)
    following = float(lines[2].split(' following')[0].replace('M', '').replace('K', '').replace(',', ''))
    if ('K' in lines[2]):
        following = int(following*1000)
    elif ('M' in lines[2]):
        following = int(following*1000000)
    else:
        following = int(following)
    
    name = lines[3]
    if len(lines)>4:
        bio = lines[4]
    else:
        bio = " "
    temp = {"Username" : username, "Followers" : followers, "Following" : following, "Name" : name, "Bio" : bio}
    organized_data.append(temp)
```

```python
import json
f = json.dumps(organized_data, indent = 3)
```
